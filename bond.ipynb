```python
from rdkit import Chem
from rdkit.Chem import AllChem
import numpy as np
smi = 'c1cccc(c1OC(=O)C)C(O)=O'
mol = Chem.MolFromSmiles(smi)
mol = Chem.AddHs(mol)
AllChem.EmbedMolecule(mol)
AllChem.MMFFOptimizeMolecule(mol)
mol = Chem.RemoveAllHs(mol)
```

```python
symbol = []
for i in range(mol.GetNumAtoms()):
    atom = mol.GetAtomWithIdx(i)
    symbol_ = atom.GetAtomicNum()
    symbol.append(symbol_)
symbol = np.array(symbol)
conf = mol.GetConformer().GetPositions()
```

```python
from utils.bond_adding import BondAdder
bond_adder = BondAdder()
rd_mol, ob_mol = bond_adder.make_mol(symbol, conf)
```
