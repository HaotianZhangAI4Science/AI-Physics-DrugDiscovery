```python
from openeye import oechem
import os.path as osp
oechem.OEThrow.SetLevel(oechem.OEErrorLevel_Error)
class args:
    query_sdf = './frag2.sdf'
    # library = './library/chemdiv.sdf'
    library = './BRD4_elaboration/frag2_zinc_searched.sdf'

qfile = oechem.oemolistream(args.query_sdf)
tfile = oechem.oemolistream(args.library)
query_name = osp.basename(args.query_sdf).split('.')[0]
library_name = osp.basename(args.library).split('.')[0]
saved_file = f'./BRD4_elaboration/{query_name}_{library_name}_searched.sdf'

# set the same aromaticity model for the query and the target file
aromodel = oechem.OEIFlavor_Generic_OEAroModelMDL
qflavor = qfile.GetFlavor(qfile.GetFormat())
qfile.SetFlavor(qfile.GetFormat(), (qflavor | aromodel))
tflavor = tfile.GetFlavor(tfile.GetFormat())
tfile.SetFlavor(tfile.GetFormat(), (tflavor | aromodel))

# read MDL query and initialize the substructure search
opts = oechem.OEMDLQueryOpts_Default | oechem.OEMDLQueryOpts_SuppressExplicitH | oechem.OEMDLQueryOpts_AddBondTopologyConstraint # | oechem.OEMDLQueryOpts_AddBondAliphaticConstraint
qmol = oechem.OEQMol()

oechem.OEReadMDLQueryFile(qfile, qmol, opts)
ss = oechem.OESubSearch(qmol)


# loop over target structures
ofs = oechem.oemolostream(saved_file)
tindex = 1
for tmol in tfile.GetOEGraphMols():
    oechem.OEPrepareSearch(tmol, ss)
    if ss.SingleMatch(tmol):
        print("hit target =", tindex)
        oechem.OEWriteMolecule(ofs, tmol)
    tindex += 1
    if tindex % 500000 == 0:
        print(f'Finished Searching {tindex} Files')


# Close the file streams
qfile.close()
tfile.close()
ofs.close()
```

```python
# Load the substructure queries from KRD4.sdf
substructure_stream = oechem.oemolistream("./frag1.sdf")

# Load the molecular library
library_stream = oechem.oemolistream("./library/IC_136982.sdf")

# Prepare to write matching molecules to a new file
ofs = oechem.oemolostream("matching_molecules.sdf")

# Iterate over each substructure query
for query_mol in substructure_stream.GetOEMols():
    # Convert OEMol to OEQMol for the query
    qmol = oechem.OEQMol()
    oechem.OEAddMols(qmol, query_mol)

    # Prepare substructure search for each query molecule
    ss = oechem.OESubSearch(qmol)

    # Reset the library stream for each new query
    library_stream.rewind()

    # Initialize a counter for progress tracking
    cnt = 0
    for mol in library_stream.GetOEMols():
        cnt += 1
        if cnt % 10000 == 0:
            print(f'Finished Searching {cnt} mols')

        if ss.SingleMatch(mol):
            # Write the matching molecule to the output file
            oechem.OEWriteMolecule(ofs, mol)

# Close the file streams
substructure_stream.close()
library_stream.close()
ofs.close()
```
